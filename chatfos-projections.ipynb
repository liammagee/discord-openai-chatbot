```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
n_years = 10
initial_sales = 10e6  # 10 million
avg_increase = 0.20  # 20%
std_dev = 0.05  # 5%

# Initialize array with initial sales
sales = np.full(n_years+1, initial_sales)

# Generate sales for each year
for i in range(1, n_years+1):
    # Generate a random percentage increase
    pct_increase = np.random.normal(avg_increase, std_dev)
    
    # Apply the percentage increase to the previous year's sales
    sales[i] = sales[i-1] * (1 + pct_increase)

# Generate array of years
years = np.arange(2023, 2023+n_years+1)

# Calculate year-on-year percentage increases
yoy_increases = (sales[1:] - sales[:-1]) / sales[:-1]

# Function to add watermark to plots
def add_watermark(ax):
    ax.text(0.5, 0.5, 'STRICTLY CONFIDENTIAL', 
            fontsize=40, color='red', alpha=0.5,
            ha='center', va='center', rotation=45,
            transform=ax.transAxes)

# Plot sales over time
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(years, sales / 1e6, marker='o')
ax.set_title('ChatFOS: 10 Year Future Sales Projections')
ax.set_xlabel('Year')
ax.set_ylabel('Sales ($M)')
ax.grid(True)
add_watermark(ax)
plt.show()

# Plot year-on-year increases over time
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(years[1:], yoy_increases * 100, marker='o')
ax.set_title('ChatFOS: Year-on-Year Sales Increases')
ax.set_xlabel('Year')
ax.set_ylabel('Sales Increase (%)')
ax.grid(True)
add_watermark(ax)
plt.show()
```
